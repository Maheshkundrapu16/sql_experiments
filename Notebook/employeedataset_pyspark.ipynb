#### library imports 

```python
from pyspark.sql.functions import *
from pyspark.sql import Window
```

#### Reading the file

```python
df = spark.read.csv(path, header=True, inferSchema=True)
```

#### aggregations

```python
from pyspark.sql.functions import *
df.groupBy("Department","Employment_Type").agg(sum("Monthly_Income").alias("monthly_sum")).display()
```

```python
df.groupBy("Employment_Type").agg(avg("Hourly_Rate").alias("average_hourlyrate"),min("Hourly_Rate").alias("min_hourlyrate"),max("Hourly_Rate").alias("max_hourlyrate")).display()
```

```python
df.groupBy("Department","Employment_Type").agg(max("Monthly_Income").alias("max_salary"),min("Monthly_Income").alias("min_salary")).orderBy("Department","Employment_Type").display()
```

#### window functions

```python
df
```

```python
window_spec = Window.partitionBy("Department").orderBy(desc("Monthly_Income"))
df.withColumn("rnk",dense_rank().over(window_spec)).select('Employee_ID','Department','Monthly_Income','rnk').filter(col("rnk")==2).display()
```

```python
window_spec = Window.partitionBy("Department").orderBy(desc("Hourly_Rate"))
df.withColumn("row_num",row_number().over(window_spec)).select('Department','Hourly_Rate','row_num')\
    .filter(col("row_num")==2).display()
```
